```python
import pandas as pd
from matplotlib import pyplot as plt

# MAE as a function of training time

graph_path = "/home/riku/projects/athlet_hr_predict/graphs"

sum_a = {'n_y':[0,5,10,15,18,20,25,30,35,40,50,60], 
        'MAE_naive':[0,1.19,2.27,3.26,3.80,4.16,5.00, 5.76,6.44,7.06,8.24,9.39],
        'MAE_model':[0,0.89,1.35,1.80,1.97,2.15,2.82,3.22,3.73
        ,4.18,5.05,6.06]}

sum_MAE = pd.DataFrame.from_dict(data=sum_a, orient='columns')
sum_MAE['MAE_ratio']=sum_MAE['MAE_model']/sum_MAE['MAE_naive']
fig, [ax1,ax2] = plt.subplots(1,2)
fig.set_size_inches(9, 2.3)
ax1.plot(sum_MAE['n_y'],sum_MAE['MAE_naive'])
ax1.plot(sum_MAE['n_y'],sum_MAE['MAE_model'])
ax2.plot(sum_MAE['n_y'],sum_MAE['MAE_ratio'])
ax2.scatter(sum_MAE['n_y'],sum_MAE['MAE_ratio'])

ax1.set_xlabel('seconds')
ax1.set_ylabel('MAE')
ax2.set_xlabel('seconds')
ax2.set_ylabel('LSTM:Naive MAE ratio')
ax1.legend(['Naive','LSTM'])
plt.tight_layout()
plt.savefig(graph_path+"/SUMMARY.png")
```
